```python
# Table 1: Sno, MovieName, Director Name, Duration, genre, ratings
import sqlite3
import csv

# Connect to the database
conn = sqlite3.connect('movies.db')
cur = conn.cursor()

# Create the table
cur.execute('''
    CREATE TABLE movies1_table1 (
        sno INTEGER PRIMARY KEY,
        movie_name TEXT,
        director1 TEXT,
        director2 TEXT,
        duration INTEGER,
        genre1 TEXT,
        genre2 TEXT,
        genre3 TEXT,
        ratings REAL
    )
''')

# Insert data from the CSV file
with open('firstCSVIMDB_merged.csv') as f:
    reader = csv.reader(f)
    next(reader) # skip header row
    for row in reader:
        cur.execute('''
            INSERT INTO movies1_table1 (sno, movie_name, director1,director2,duration,ratings,genre1,genre2,genre3)
            VALUES (?, ?, ?, ?, ?, ?,?, ?, ?)
        ''', (row[0], row[1], row[6], row[7], row[2], row[4], row[8], row[9], row[10]))

    
# Commit changes and close the connection
conn.commit()
conn.close()
```

```python
# Table 2: Movie Name, stars, votes, Genre, Gross collection, popularity, Certification
import sqlite3
import csv

# Connect to the database
conn = sqlite3.connect('movies.db')
cur = conn.cursor()

# Create the table
cur.execute('''
    CREATE TABLE movies2_table2 (
        sno INTEGER PRIMARY KEY,
        movie_name TEXT,
        star1 REAL,
        star2 REAL,
        star3 REAL,
        star4 REAL,
         votes INTEGER,
        genre1 TEXT,
        genre2 TEXT,
        genre3 TEXT,
        gross_collection REAL,
        popularity INTEGER,
        Certification REAL      
                 
    )
''')

# 
# 
# Insert data from the CSV file
with open('secondCSVIMDB.csv' ,'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader) # skip header row
    for row in reader:
        cur.execute('''
            INSERT INTO movies2_table2 (sno, movie_name, star1,star2,star3,star4,votes,genre1,genre2,genre3,gross_collection,popularity,Certification)
            VALUES (?, ?, ?, ?, ?, ?,?, ?, ?, ?, ?, ?,?)
        ''', (row[0], row[1], row[6], row[7], row[8], row[9], row[2], row[10], row[11], row[12],row[3], row[4], row[5]))

    
# Commit changes and close the connection
conn.commit()
conn.close()
```

```python
# 1) Display all details from both tables where movie names are the same.
import sqlite3
# Connect to the database
conn = sqlite3.connect('movies.db')
cur = conn.cursor()

# #Query from first Table
# cur.execute('select * from movies1_table1')

# Query from Second Table
cur.execute("SELECT * FROM movies1_table1 INNER JOIN movies2_table2 ON movies1_table1.movie_name = movies2_table2.movie_name")

# Fetch all records
# as list of tuples
records2 = cur.fetchall()
 
# Display result 
for row2 in records2:
    # show row
    print(row2)
conn.close()
```

```python
# 2)Display all movie names, Director, ratings, and gross collection where the genre is action.
# Connect to the database
conn = sqlite3.connect('movies.db')
cur = conn.cursor()


# # Query from Second Table
cur.execute("select m1.movie_name,m1.director1,m1.director2,m1.ratings,m2.gross_collection from movies1_table1 m1 JOIN movies2_table2 m2 ON m1.movie_name = m2.movie_name WHERE m2.genre1 ='Action' OR m2.genre2 ='Action' OR m2.genre3 ='Action'")

# Fetch all records
# as list of tuples
records2 = cur.fetchall()
 
# Display result 
for row2 in records2:
    # show row
    print(row2)
conn.close()
```

```python
# 3)Display all details from both tables with the highest gross collection.

# Connect to the database
conn = sqlite3.connect('movies.db')
cur = conn.cursor()


# # Query from Second Table
cur.execute("select * from movies1_table1 JOIN movies2_table2 on movies1_table1.movie_name = movies2_table2.movie_name WHERE movies2_table2.gross_collection = (select MAX(gross_collection) from movies2_table2)")

# Fetch all records
# as list of tuples
records2 = cur.fetchall()
 
# Display result 
for row2 in records2:
    # show row
    print(row2)
conn.close()
```
