```python
# 1)Display all details from both tables where movie names are the same.
# merge the two dataframes on the 'movie name' column
merged_df = pd.merge(IMDB1df, IMDB2df, on='movie', how='inner')

# display the merged dataframe
merged_df
```

```python
# 2)Display all movie names, Director, ratings, and gross collection where the genre is action.
merged_df = pd.merge(IMDB1df, IMDB2df, on='movie', how='inner')

# select rows with action genres
filter_genre = [ 'Action']

selected_rows = merged_df[merged_df['genre1'].isin(filter_genre) |  merged_df['genre2'].isin(filter_genre) | merged_df['genre3'].isin(filter_genre)]

# display all details of selected rows
selected_rows[['movie','Director1','Director2','rating','gross_collection']]
```

```python
# 3)Display all details from both tables with the highest gross collection.
joined_table = pd.merge(IMDB1df, IMDB2df, on='movie')

# filter the joined table based on the highest gross collection
highest_gross = joined_table['gross_collection'].max()
result = joined_table[joined_table['gross_collection'] == highest_gross]

# display the resulting DataFrame
result
```

```python
# 4) Display all details from both tables with the highest ratings
joined_table = pd.merge(IMDB1df, IMDB2df, on='movie')

# filter the joined table based on the highest ratings
highest_gross = joined_table['rating'].max()
result = joined_table[joined_table['rating'] == highest_gross]

# display the resulting DataFrame
result
```

```python
# 5)Display all details from both tables with the lowest gross collection and lowest ratings

merged_df = pd.merge(IMDB1df, IMDB2df, on='movie')

# get rows with the lowest gross collection and lowest ratings
lowest_gross = merged_df[merged_df['gross_collection'] == merged_df['gross_collection'].min()]
lowest_rating = merged_df[merged_df['rating'] == merged_df['rating'].min()]

# display all details from both tables with the lowest gross collection and lowest ratings
result = pd.concat([lowest_gross, lowest_rating], axis=0)
result
```
